# Cluster the pickup time of the day into 30-minute intervals (e.g. from 02:00 to 02:30)

1. ho trasformato i valori in data
2. ho trasformato la data in secondi
3. ho creato i bins come intervalli di di 30 minuti
4. ho creato le labels dei singoli intervalli di 30 minuti
5. ho assegnato ad ogni orario l'intervallo corrispondente con la funzione cut

```python

df_clean['pickup_time_cluster'] = pd.to_datetime(df_clean['tpep_pickup_datetime'])
df_clean['pickup_time_cluster'] = (df_clean['pickup_time_cluster'].dt.hour * 3600 + df_clean['pickup_time_cluster'].dt.minute * 60 + df_clean['pickup_time_cluster'].dt.second)
```

```python
bins = [i * 1800 for i in range(49)]
bins_time = [f"{(secondi // 3600):02}:{((secondi % 3600) // 60):02}" for secondi in bins]
labels = [f"{bins_time[i]} - {bins_time[i+1]}" for i in range(len(bins_time)-1)]

df_clean['pickup_time_cluster'] = pd.cut(df_clean['pickup_time_cluster'], bins=bins, labels=labels)
df_clean
```

# Build chains of trips. Two trips are consecutive in a chain if (a) they have the same VendorID, (b) the pickup location of the second trip is also the dropoff location of the first trip, (c) the pickup time of the second trip is after the dropoff time of the first trip, and (d) the pickup time of the second trip is at most 2 minutes later than the dropoff time of the first trip.

```python
import pandas as pd
from tqdm import tqdm  


df_clean['tpep_pickup_datetime'] = pd.to_datetime(df_clean['tpep_pickup_datetime'])
df_clean['tpep_dropoff_datetime'] = pd.to_datetime(df_clean['tpep_dropoff_datetime'])


df_clean = df_clean.sort_values(by=['VendorID', 'tpep_pickup_datetime'])


chain_values = [None] * len(df_clean) 


chain_number = 1


for i in tqdm(range(1, len(df_clean)), desc="Processing trips", ncols=100):
    row_i = df_clean.iloc[i]
    row_prev = df_clean.iloc[i - 1]
    

    if (row_i['VendorID'] == row_prev['VendorID'] and 
        row_i['DOLocationID'] == row_prev['PULocationID'] and 
        (row_i['tpep_pickup_datetime'] - row_prev['tpep_dropoff_datetime']).total_seconds() <= 120):
        

        chain_values[i] = chain_number
    else:

        chain_number += 1
        chain_values[i] = chain_number


df_clean['chain'] = chain_values

```
