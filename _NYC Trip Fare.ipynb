# Cluster the pickup time of the day into 30-minute intervals (e.g. from 02:00 to 02:30)

1. abbiamo tradormato i valori in data
2. abbiamo trasformato la data in secondi
3. abbiamo creato i bins come intervalli di di 30 minuti
4. abbiamo creato le labels dei singoli intervalli di 30 minuti
5. abbiamo assegnato ad ogni orario l'intervallo corrispondente con la funzione cut

```python

df_clean['pickup_time_cluster'] = pd.to_datetime(df_clean['tpep_pickup_datetime'])
df_clean['pickup_time_cluster'] = (df_clean['pickup_time_cluster'].dt.hour * 3600 + df_clean['pickup_time_cluster'].dt.minute * 60 + df_clean['pickup_time_cluster'].dt.second)
```

```python
bins = [i * 1800 for i in range(49)]
bins_time = [f"{(secondi // 3600):02}:{((secondi % 3600) // 60):02}" for secondi in bins]
labels = [f"{bins_time[i]} - {bins_time[i+1]}" for i in range(len(bins_time)-1)]

df_clean['pickup_time_cluster'] = pd.cut(df_clean['pickup_time_cluster'], bins=bins, labels=labels)
df_clean
```

```python
import pandas as pd

# 1) Prepara i datetime
df = df_clean.copy().reset_index(drop=True)
df['tpep_pickup_datetime']  = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])

n = len(df)
df['idx'] = range(n)

# 2) Costruisci le tabelle "left" (dropoff) e "right" (pickup) con gli indici originali
left_cols  = ['idx','VendorID','DOLocationID','tpep_dropoff_datetime']
right_cols = ['idx','VendorID','PULocationID','tpep_pickup_datetime']

# 3) Per ogni gruppo (VendorID, DOLocationID) esegui merge_asof
matches_list = []
for (vid, dol), grp in df.groupby(['VendorID','DOLocationID'], sort=False):
    # left: dropoff
    left_grp = (
        grp[left_cols]
        .rename(columns={'idx':'i','tpep_dropoff_datetime':'dropoff_time'})
        .sort_values('dropoff_time')
        .reset_index(drop=True)
    )
    # right: pickup (rinomino PULocationID in DOLocationID per match)
    right_grp = (
        grp[right_cols]
        .rename(columns={'idx':'j','PULocationID':'DOLocationID','tpep_pickup_datetime':'pickup_time'})
        .sort_values('pickup_time')
        .reset_index(drop=True)
    )
    # merge_asof entro 120s
    m = pd.merge_asof(
        left_grp, right_grp,
        left_on='dropoff_time', right_on='pickup_time',
        direction='forward',
        tolerance=pd.Timedelta(seconds=120)
    )
    matches_list.append(m)

# 4) Unisci i risultati e prendi solo le coppie valide
matches = pd.concat(matches_list, ignore_index=True)
matches = matches.dropna(subset=['j'])
pairs = list(zip(matches['i'].astype(int), matches['j'].astype(int)))

# 5) Union-Find (DSU) per aggregare gli indici in catene
parent = list(range(n))
def find(x):
    while parent[x] != x:
        parent[x] = parent[parent[x]]
        x = parent[x]
    return x

def union(a, b):
    ra, rb = find(a), find(b)
    if ra != rb:
        parent[rb] = ra

for a, b in pairs:
    union(a, b)

# 6) Assegna un ID di chain a ogni root
chain_map = {}
next_id   = 1
chain     = [None] * n

for idx in range(n):
    r = find(idx)
    if r not in chain_map:
        chain_map[r] = next_id
        next_id += 1
    chain[idx] = chain_map[r]

# 7) Aggiungi la colonna chain al DataFrame originale
df_clean['chain'] = chain

df_clean




```
